```python
#get dummies 
df1=pd.get_dummies(df[["sex","smoker","day","time"]],drop_first=True)
df1
```

```python

dfb=pd.concat([df,df1],axis=1) #combination 2 datas
dfb
```

```python
#dropping columns
dfb.drop(columns=["sex","time","smoker","day"],inplace=True)
```

```python
#x and y variable seperation
x=dfb.drop(columns=["tip"])
x
```

```python
y=dfb["tip"]
y
```

```python
#trainig data testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
#MAE percentage
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

```python
#r2 score
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
